```python
import pandas as pd
import matplotlib.pyplot as plt

dataset_path = 'datasets\Dataset salary 2024.csv'
df = pd.read_csv(dataset_path)

df = df.dropna()

avg_salary_by_job = df.groupby('job_title')['salary_in_usd'].mean().sort_values()

plt.figure(figsize=(12, 30))
avg_salary_by_job.plot(kind='barh', color='skyblue')
plt.title('Average Salary by Job Title')
plt.xlabel('Average Salary in USD')
plt.ylabel('Job Title')
plt.show()

```

```python
# Filter job titles where mean salary is > 300,000 USD
avg_salary_by_job = avg_salary_by_job[avg_salary_by_job > 200000]

# Plotting average salary for each job title
plt.figure(figsize=(50, 12))
avg_salary_by_job.plot(kind='barh', color='skyblue')
plt.title('Average Salary by Job Title (Salary > 200,000 USD)')
plt.xlabel('Job Title')
plt.ylabel('Average Salary in USD')
plt.show()
```
